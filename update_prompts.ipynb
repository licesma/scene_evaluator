```python
import yaml
import os
from pathlib import Path

# Base directory for videos
video_base_dir = "/data2/openreal2sim/outputs/videos"

# Update metadata.yaml for each video with its corresponding prompt
for name, prompt in zip(names, prompts):
    metadata_path = os.path.join(video_base_dir, name, "metadata.yaml")
    
    # Check if metadata file exists
    if os.path.exists(metadata_path):
        # Read existing metadata
        with open(metadata_path, 'r') as f:
            metadata = yaml.safe_load(f)
        
        # Add/update the prompt field
        if metadata is None:
            metadata = {}
        metadata['prompt'] = prompt
        
        # Write updated metadata
        with open(metadata_path, 'w') as f:
            yaml.dump(metadata, f, default_flow_style=False)
        
        print(f"Updated: {name}")
    else:
        print(f"Not found: {metadata_path}")

print("\nDone! All metadata files have been updated with prompts.")
```
